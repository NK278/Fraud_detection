```python
import csv

# Function to upload data in batches
def upload_data_in_batches(collection, data, batch_size=1000):
    total_records = len(data)
    num_batches = (total_records + batch_size - 1) // batch_size  # Calculate number of batches
    for i in range(num_batches):
        start_idx = i * batch_size
        end_idx = min((i + 1) * batch_size, total_records)
        batch_data = data[start_idx:end_idx]
        collection.insert_many(batch_data)
        print(f"Batch {i+1}/{num_batches} uploaded successfully")

# MongoDB database and collection
db = client['Fraud_Detection2']
collection = db['Project_SML1']

# CSV file path
csv_file_path = 'data\FraudTrain.csv'

# Read CSV file and upload data in batches
with open(csv_file_path, 'r') as file:
    reader = csv.DictReader(file)
    data = [row for row in reader]

upload_data_in_batches(collection, data)
print("All data uploaded successfully to the new database in MongoDB.")
```
